```python
# ----------------------------
# DAY 1: PYTHON BASICS
# ----------------------------

print("üî• Welcome to Day 1: Python Basics üî•\n")

# 1Ô∏è‚É£ Variables & Data Types
print("== Variables & Data Types ==")
x = 10               # Integer
y = 3.14             # Float
name = "Hridya"      # String
is_student = True    # Boolean

print("Integer:", x)
print("Float:", y)
print("String:", name)
print("Boolean:", is_student)

# 2Ô∏è‚É£ Basic Math Operations
print("\n== Basic Math Operations ==")
a = 15
b = 4

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Exponent:", a ** b)
print("Modulo:", a % b)

# 3Ô∏è‚É£ Strings
print("\n== Strings ==")
greeting = "Hello"
message = greeting + " " + name
print(message)
print("String Length:", len(message))
print("Uppercase:", message.upper())
print("Lowercase:", message.lower())

# 4Ô∏è‚É£ Lists
print("\n== Lists ==")
fruits = ["apple", "banana", "mango"]
print("Fruits List:", fruits)
print("First fruit:", fruits[0])

# Adding/Removing
fruits.append("orange")
fruits.remove("banana")
print("Updated Fruits List:", fruits)

# 5Ô∏è‚É£ Loops
print("\n== Loops ==")
print("For Loop:")
for fruit in fruits:
    print(fruit)

print("\nWhile Loop:")
count = 0
while count < 3:
    print("Count:", count)
    count += 1

# 6Ô∏è‚É£ Functions
print("\n== Functions ==")
def greet(person):
    return f"Hello {person}, welcome to Python!"

print(greet("Hridya"))

# 7Ô∏è‚É£ Mini Exercises
print("\n== Mini Exercises ==")

# Exercise 1: Add two numbers
def add_numbers(a, b):
    return a + b

print("Exercise 1:", add_numbers(5, 7))

# Exercise 2: Check if number is even or odd
def even_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

print("Exercise 2:", even_odd(10))
print("Exercise 2:", even_odd(7))

# Exercise 3: Print all numbers from 1 to 10
print("Exercise 3: Numbers 1 to 10")
for i in range(1, 11):
    print(i, end=" ")
print("\n")

# Exercise 4: Create a list of squares from 1 to 5
squares = [i**2 for i in range(1,6)]
print("Exercise 4: Squares List", squares)

```
