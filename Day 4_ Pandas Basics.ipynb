```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Dataset loaded successfully!")
```

```python
print("Dataset info:")
df.info()
```

```python
# Select single column
df['Age'].head()
```

```python
# Select multiple columns
df[['Name', 'Age', 'Sex']].head()
```

```python
# Select rows using index
df.iloc[0:5]
```

```python
# Select rows using condition
df[df['Age'] > 30].head()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Fill missing Age with mean
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

print("Missing values handled!")
```

Basic Data Analysis

```python
# Survival count
df['Survived'].value_counts()
```

```python
# Survival by gender
df.groupby('Sex')['Survived'].mean()
```

```python
# Survival by passenger class
df.groupby('Pclass')['Survived'].mean()
```

Create New Features

```python
# Create age groups
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[0, 12, 20, 40, 60, 100],
    labels=['Child', 'Teen', 'Adult', 'Middle-Aged', 'Senior']
)

df[['Age', 'AgeGroup']].head()
```

Sorting Data

```python
# Sort by age
df.sort_values(by='Age', ascending=False).head()
```

Mini Exercises

```python
print("== Mini Exercises ==")

# Exercise 1: Average age of passengers
print("Average Age:", df['Age'].mean())

# Exercise 2: Count passengers in each class
print("Passengers per class:")
print(df['Pclass'].value_counts())

# Exercise 3: Survival rate by AgeGroup
print("Survival rate by AgeGroup:")
print(df.groupby('AgeGroup')['Survived'].mean())

# Exercise 4: Female survival rate
female_survival = df[df['Sex'] == 'female']['Survived'].mean()
print("Female Survival Rate:", female_survival)
```

Save Cleaned Data

```python
# Save cleaned dataset
df.to_csv("titanic_cleaned.csv", index=False)

print("Cleaned dataset saved as titanic_cleaned.csv")
```

Survival rate by Embarked port

```python
# Survival rate by Embarked port
survival_by_embarked = df.groupby('Embarked')['Survived'].mean()
print("Survival Rate by Embarked Port:")
print(survival_by_embarked)
```

Average fare per Passenger Class

```python
# Average fare per passenger class
avg_fare_by_class = df.groupby('Pclass')['Fare'].mean()
print("\nAverage Fare per Passenger Class:")
print(avg_fare_by_class)
```

Create new column: FamilySize

```python
# Create FamilySize column
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Preview result
df[['SibSp', 'Parch', 'FamilySize']].head()
```
