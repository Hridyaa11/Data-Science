```python
# ----------------------------------
# DAY 6: DATA VISUALIZATION
# ----------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("ðŸŽ¨ Day 6: Data Visualization with Matplotlib & Seaborn ðŸŽ¨\n")

# Load dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Basic cleaning
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Create FamilySize
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

Bar Plot â€“ Survival Count

```python
plt.figure(figsize=(6,4))
df['Survived'].value_counts().plot(kind='bar')
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

Count Plot â€“ Survival by Gender

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

Bar Plot â€“ Survival Rate by Passenger Class

```python
survival_by_class = df.groupby('Pclass')['Survived'].mean()

plt.figure(figsize=(6,4))
survival_by_class.plot(kind='bar')
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

Histogram â€“ Age Distribution

```python
plt.figure(figsize=(6,4))
plt.hist(df['Age'], bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Box Plot â€“ Fare by Passenger Class

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

Violin Plot â€“ Age vs Survival

```python
plt.figure(figsize=(6,4))
sns.violinplot(x='Survived', y='Age', data=df)
plt.title("Age Distribution by Survival")
plt.show()
```

Scatter Plot â€“ Age vs Fare

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Age'], df['Fare'])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

Heatmap â€“ Correlation Matrix

```python
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

Pair Plot (Quick EDA)

```python
sns.pairplot(df[['Age', 'Fare', 'Survived', 'Pclass']])
plt.show()
```

Survival Rate by Family Size (Line Plot)

```python
family_survival = df.groupby('FamilySize')['Survived'].mean()

plt.figure(figsize=(6,4))
family_survival.plot(marker='o')
plt.title("Survival Rate by Family Size")
plt.xlabel("Family Size")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# Make sure df is already loaded and cleaned
# df should contain: Age, Fare, Embarked, Sex, FamilySize, Survived

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1ï¸âƒ£ Bar Chart â€“ Survival Rate by Embarked Port
survival_by_embarked = df.groupby('Embarked')['Survived'].mean()

plt.figure(figsize=(6,4))
survival_by_embarked.plot(kind='bar')
plt.title("Survival Rate by Embarked Port")
plt.xlabel("Embarked Port (C, Q, S)")
plt.ylabel("Survival Rate")
plt.show()


# 2ï¸âƒ£ Histogram â€“ Fare Distribution
plt.figure(figsize=(6,4))
plt.hist(df['Fare'], bins=30)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()


# 3ï¸âƒ£ Boxplot â€“ Age by Gender
plt.figure(figsize=(6,4))
sns.boxplot(x='Sex', y='Age', data=df)
plt.title("Age Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()


# 4ï¸âƒ£ Heatmap â€“ Only Age, Fare, FamilySize, Survived
selected_features = df[['Age', 'Fare', 'FamilySize', 'Survived']]
correlation = selected_features.corr()

plt.figure(figsize=(6,5))
sns.heatmap(correlation, annot=True)
plt.title("Correlation Heatmap (Age, Fare, FamilySize, Survived)")
plt.show()
```

What each plot shows you:

Bar chart â†’ Which port had the highest survival rate

Histogram â†’ How ticket prices are distributed (most were cheap, few were expensive)

Boxplot â†’ Age differences between males and females + outliers

Heatmap â†’ How strongly Age, Fare, FamilySize relate to Survival
